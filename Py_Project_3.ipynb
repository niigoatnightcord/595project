```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x).
def f(x):
    return np.sqrt(x**2 + 5)


# Plot the function f(x), np.linspace might be helpful.
x_vals = np.linspace(-5, 5, 100)
y_vals = f(x_vals)
plt.plot(x_vals, y_vals)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x)')
plt.grid(True)
plt.show()
```

```python
# Define the derivative of the function f(x), for gradient descent
def df(x):
    return x / np.sqrt(x**2 + 5)

# write a function to run the Gradient Descent Algorithm, take input arguments x_0, step_size, and num_iterations; return a list/array of x_k values.
def gradient_descent(x_0, step_size, num_iterations):
    x_k = x_0
    x_history = [x_0]
    for _ in range(num_iterations):
        x_k = x_k - step_size * df(x_k)
        x_history.append(x_k)
    return x_history
# set up/assign initial values
x_0 = 2.4
step_size = 1
num_iterations = 50

# Run gradient descent by calling the function
x_history = gradient_descent(x_0, step_size, num_iterations)
x_T = x_history[-1]

# report x_T by printing out the value
print("x_T:", x_T)


# Plot the f(x_k) values vs. iteration number
plt.plot(range(num_iterations + 1), [f(x) for x in x_history])
plt.xlabel('Iteration')
plt.ylabel('f(x_k)')
plt.title('f(x_k) vs. Iteration')
plt.grid(True)
plt.show()
```

```python
# create a list containing step sizes
step_sizes = [5, 3, 1, 0.5]

# initialize a list to store the results from different step sizes
results = []

# loop through the step sizes and perform GD and record results
for step_size in step_sizes:
    x_history = gradient_descent(x_0, step_size, num_iterations)
    results.append(-np.diff([f(x) for x in x_history]))


# plot multiple curves in a single graph
plt.plot(range(1, num_iterations + 1), results[0], label='step_size=5')
plt.plot(range(1, num_iterations + 1), results[1], label='step_size=3')
plt.plot(range(1, num_iterations + 1), results[2], label='step_size=1')
plt.plot(range(1, num_iterations + 1), results[3], label='step_size=0.5')
plt.xlabel('Iteration')
plt.ylabel('f(x_{k-1}) - f(x_k)')
plt.title('f(x_{k-1}) - f(x_k) vs. Iteration for Different Step Sizes')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate synthetic data
np.random.seed(17)   # change the seed to be the last two digits of your SB ID.
X = 2 * np.random.rand(100, 2)  # Generate 100 random 2D data points
y = 4 + 4 * X[:, 0] + 0.5 * X[:, 1] + np.random.randn(100)  # Generate y values with noise

# Add a column of ones to X for the bias terms (intercept) in the normal equation
X_b = np.c_[np.ones((100, 1)), X]


# Find theta using the normal equation
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)



# Print the intercept and coefficients of the regression plane
print("Intercept:", theta_best[0])
print("Coefficients:", theta_best[1:])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the generated data and the regression plane
ax.scatter(X[:, 0], X[:, 1], y, c='b', marker='o', label='Data points')
x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
x1, x2 = np.meshgrid(np.linspace(x1_min, x1_max, 20), np.linspace(x2_min, x2_max, 20))
y_pred = theta_best[0] + theta_best[1] * x1 + theta_best[2] * x2
ax.plot_surface(x1, x2, y_pred, alpha=0.5, color='r', label='Regression plane')

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('y')
ax.legend()
plt.show()
```

## 4. Logistic Regression with Numpy and Graphing with Matplotlib (20 pts)

In class, we implemented a logistic regression model to perform binary classification. Here, I provided the exact same code that we discussed in class, but the values for learning_rate and num_epochs are modified. I also removed all the comments. You task is to
1. Tune the Learning Rate and Number of Epochs: Your first task is to fine-tune the values for the learning_rate and num_epochs parameters. Your goal is to identify suitable values that will enable us to converge to a set of parameter values $\theta$ closely approximating the optimal parameter values $\theta^*$.

By accomplishing these tasks, we aim to achieve a better understanding of the logistic regression model's behavior and its parameter optimization process.

```python
import numpy as np
import matplotlib.pyplot as plt
#Sigmoid function (used for logistic regression predictions)
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

#random seed for reproducibility
np.random.seed(595)
X = np.random.rand(2, 100)#generate synthetic data
w_true = np.array([1.5, -2.5])#true parameters for logistic regression
b_true = 1.0  # True bias
probabilities = sigmoid(np.dot(w_true.T, X) + b_true)
Y = (probabilities > 0.5).astype(int)#generate labels by thresholding probabilities at 0.5
X = X + 0.3 * np.random.rand(2, 100) - 0.1 * np.random.rand(2, 100)#add random noise to the data
#split data into training and testing sets (80% training, 20% testing)
X_train, X_test = X[:,:80], X[:,80:]
Y_train, Y_test = Y[:80], Y[80:]

#initialize weights and bias to zero
w = np.zeros(X.shape[0])
b = 0.0
learning_rate = 14.85 #hyperparameters from 12.83 to 14.84
num_epochs = 600
#training loop (gradient descent)
for epoch in range(num_epochs):
    A_train = sigmoid(np.dot(w.T, X_train) + b)#model's predictions on the training data

    #gradients of the loss function with respect to weights and bias
    dJdw = np.dot(X_train, (A_train - Y_train).T) / len(Y_train)
    dJdb = np.mean(A_train - Y_train)

    w -= learning_rate * dJdw#update the weights and bias using gradient descent
    b -= learning_rate * dJdb

A_train = sigmoid(np.dot(w.T, X_train) + b)#compute the final predictions on the training set
predictions_train = (A_train > 0.5).astype(int)

A_test = sigmoid(np.dot(w.T, X_test) + b)#compute the final predictions on the test set
predictions_test = (A_test > 0.5).astype(int)

train_accuracy = np.mean(predictions_train == Y_train)#accuracy on the training and test sets

test_accuracy = np.mean(predictions_test == Y_test)

print(f"Training Set Accuracy: {train_accuracy:.2f}")
print(f"Test Set Accuracy: {test_accuracy:.2f}")

prev_loss = float('inf')  # Initialize with a very large value
for epoch in range(num_epochs):
    A_train = sigmoid(np.dot(w.T, X_train) + b)
    # Compute the loss (e.g., binary cross-entropy)
    loss = -np.mean(Y_train * np.log(A_train) + (1 - Y_train) * np.log(1 - A_train))

    # Check for convergence (if loss change is below a small threshold)
    if abs(prev_loss - loss) < 1e-5:
        print(f"Converged at epoch {epoch + 1}, Loss: {loss:.4f}")
        break
    prev_loss = loss
```

```python
#visualize the decision boundary for the training set
plt.scatter(X[0, :80], X[1,:80], c=Y[:80], cmap=plt.cm.Paired)
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)
ax = plt.gca()

xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))
Z = np.dot(w.T, np.c_[xx.ravel(), yy.ravel()].T) + b
Z = sigmoid(Z)#sigmoid to get probabilities
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, colors='k', levels=[0.5], alpha=0.5, linestyles=['--'])#plot the decision boundary

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Decision Boundary')
plt.show()
```

```python
#decision boundary for the test set
plt.scatter(X[0, 80:], X[1, 80:], c=Y[80:], cmap=plt.cm.Paired)
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)

ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))
Z = np.dot(w.T, np.c_[xx.ravel(), yy.ravel()].T) + b
Z = sigmoid(Z)#sigmoid to get probabilities
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, colors='k', levels=[0.5], alpha=0.5, linestyles=['--'])
# plot the decision boundary

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Decision Boundary')
plt.show()
```
